```python
import pandas as pd
```

```python
data_latih = pd.read_csv('../output/CLEAN-dataset_mobil_listrik.csv', encoding='utf-8')
data_latih
```

```python
data_uji = pd.read_csv('../output/data-clean-ind-new.csv', encoding='utf-8')
data_uji
```

```python
X_train = data_latih['resultStemming']
y_train = data_latih['sentimen']

X_test = data_uji['resultStemming']
```

### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
X_train_tfidf = vectorizer.fit_transform(X_train)

word_weight_list = list(zip(vectorizer.get_feature_names_out(), vectorizer.idf_))
word_weight_list.sort(key=lambda x: x[1])

for word, idf_weight in word_weight_list:
    print(f"{word:15s}: {idf_weight:.3f}")
```

```python
X_test_tfidf = vectorizer.transform(X_test)

word_weight_list = list(zip(vectorizer.get_feature_names_out(), vectorizer.idf_))
word_weight_list.sort(key=lambda x: x[1])

for word, idf_weight in word_weight_list:
    print(f"{word:15s}: {idf_weight:.3f}")
```

### Make New DataFrame

```python
df_resultTfIdf = pd.DataFrame(X_train_tfidf.toarray(), columns=vectorizer.get_feature_names_out())
df_resultTfIdf['Sentiment'] = y_train.values
```

```python
df_resultTfIdf
```

```python
df_resultTestTfIdf = pd.DataFrame(X_test_tfidf.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
df_resultTestTfIdf
```

### Export Result

```python
df_resultTfIdf.to_csv(
    '../tfidf/TFIDF-TRAIN-dataset_mobil_listrik.csv', 
    index=False
)
```

```python
df_resultTestTfIdf.to_csv(
    '../tfidf/TFIDF-TEST-data-clean-ind-new.csv', 
    index=False
)
```
