```python
import pandas as pd
```

### Merge Dataset

```python
df1 = pd.read_csv('../dataset/dataset_mobil_listrik.csv', encoding='utf-8')
df2 = pd.read_csv('../dataset/tweets_sentiment_covid_fix.csv', encoding='utf-8')
df3 = pd.read_csv('../dataset/dataset_digital_bank_indo.csv', encoding='utf-8')
```

```python
df1[['text_cleaning', 'sentimen']]
```

```python
df2[['text_clean', 'polarity']]
```

```python
df3[['text_clean', 'sentimen']]
```

```python
# Transform label df2

def translate_polarity(polarity):
    if polarity == 'positive':
        return 'positif'
    elif polarity == 'negative':
        return 'negatif'
    else:
        return 'netral'
```

```python
df2['sentimen'] = df2['polarity'].apply(translate_polarity)
df2[['text_clean', 'polarity', 'sentimen']].head(150)
```

```python
df1 = df1.rename(columns={'text_cleaning': 'text'})
df2 = df2.rename(columns={'text_clean': 'text'})
df3 = df3.rename(columns={'text_clean': 'text'})
```

```python
df_merge = pd.concat([df1[['text', 'sentimen']], df2[['text', 'sentimen']].head(150), df3[['text', 'sentimen']]], ignore_index=True)
df_merge
```

```python
df_merge.to_csv(
    '../output/merge/dataset_merge.csv',
    columns=df_merge.columns,
    index=False
)
```
