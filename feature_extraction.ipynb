```python
import pandas as pd
```

```python
df = pd.read_csv('./output/data-clean-ind-test.csv', encoding='utf-8')
df
```

### Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['resultStemming'], df[' label '], test_size=0.2, random_state=42)
```

### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

#### Data Latih

```python
vectorizer = TfidfVectorizer()
```

```python
X_train_tfidf = vectorizer.fit_transform(X_train)

print("Kata: Skor IDF (Data Latih)")
for word, idf in zip(vectorizer.get_feature_names_out(), vectorizer.idf_):
    print(f"{word}: {idf}")
```

```python
df_resultLatihTfIdf = pd.DataFrame(X_train_tfidf.toarray(), columns=vectorizer.get_feature_names_out())
df_resultLatihTfIdf
```

```python
df_resultLatihTfIdf['label'] = y_train.values
df_resultLatihTfIdf.to_csv('./tfidf/result_latih_tfidf.csv', index=False)
```

#### Data Uji

```python
X_test_tfidf = vectorizer.transform(X_test)

print("\nKata: Skor IDF (Data Uji)")
for word, idf in zip(vectorizer.get_feature_names_out(), vectorizer.idf_):
    print(f"{word}: {idf}")
```

```python
df_resultUjiTfIdf = pd.DataFrame(X_test_tfidf.toarray(), columns=vectorizer.get_feature_names_out())
df_resultUjiTfIdf
```

```python
df_resultUjiTfIdf['label'] = y_test.values
df_resultUjiTfIdf.to_csv('./tfidf/result_uji_tfidf.csv', index=False)
```
